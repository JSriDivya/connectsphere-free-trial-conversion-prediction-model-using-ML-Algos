## The Challenge We're Solving (Problem Statement)

Many people sign up for ConnectSphere's free trial, but a good number don't end up paying for a subscription. We're currently not very targeted in how we talk to these trial users, which means we might be missing chances to convert interested users and not using our sales and marketing time wisely.

> Our goal is to use data from a user's first week to predict who is likely to buy a subscription, so we can focus our efforts and increase our number of paying customers.

---

## What We Looked At (The Data)

We focused on what users did in their first 7 days:
*   How many projects they started.
*   If they assigned tasks to others.
*   How many team members they invited.
*   If they connected other tools (integrations).
*   A general "activity score."
*   If they finished our startup tutorial.
*   How they found us (e.g., Google search, an ad, a friend's referral).

And, of course, whether they eventually **paid for a subscription** after their 14-day trial.

We also looked at which user actions our best model thought were most important for predicting a sale.

# **STEP 2: DATA GENERATION (SIMULATED)**

```python
# --- STEP 2: DATA GENERATION (SIMULATED) ---
import numpy as np
import pandas as pd

#1> First define Campaign Channels
#2> Users coming thorugh diff channels
#3> Once successfully onboards users may create projects
#4> post creation user will invite diff team members
#5> Tasks are assigned
#6> what and how are they integrating projects - google drive/ one drive etc
#7> Daily activity score / time spend on SaaS platform

#--------------------------------------------------------------------------------------------------

def generate_data(num_users=2500):
    data = []

    for i in range(num_users):
        user_id = 1000 + i  # Assign a unique user ID starting from 1000
#--------------------------------------------------------------------------------------------------
        # Randomly assign a marketing channel with defined probabilities
        channel = np.random.choice(['Organic', 'Paid_Ad', 'Referral', 'Social_Media'],
                                   p=[0.35, 0.30, 0.20, 0.15])
#--------------------------------------------------------------------------------------------------
        # Set a base conversion probability for every user
        prob = 0.10

        # Modify conversion probability based on marketing channel
        prob += {
            'Referral': 0.12,      # High trust = higher chance to convert
            'Organic': 0.05,       # Moderate intent = medium impact
            'Paid_Ad': -0.03       # Lower quality = negative impact
        }.get(channel, 0)
#--------------------------------------------------------------------------------------------------
        # Simulate if the user completed the onboarding tutorial (65% chance)
        onboarding = np.random.choice([0, 1], p=[0.35, 0.65])
        if onboarding:
            prob += 0.15  # Completing onboarding boosts conversion likelihood
#--------------------------------------------------------------------------------------------------
        # Simulate number of projects created (more if onboarding is completed)
        projects = np.random.poisson(2.0 if onboarding else 1.0)
        if projects > 0:
            prob += projects * 0.06  # Each project adds 6%
        if projects > 3:
            prob += 0.05  # Bonus for high activity (more than 3 projects)
#--------------------------------------------------------------------------------------------------
        # Simulate number of team members invited
        invites = np.random.poisson(0.3 + 0.25 * projects) if projects > 0 else 0
        if invites > 0:
            prob += invites * 0.05  # 5% boost per invite
        if invites > 2:
            prob += 0.04  # Extra bonus for more than 2 invites
#--------------------------------------------------------------------------------------------------
        # Simulate tasks assigned to others
        tasks = np.random.poisson(invites * 1.0 + projects * 0.5) if invites > 0 else 0
        if tasks > 1:
            prob += 0.03  # Light boost for some task delegation
        if tasks > 5:
            prob += 0.05  # Extra boost for high task delegation
#--------------------------------------------------------------------------------------------------
        # Simulate integrations used (Slack, Google Drive, etc.)
        integrations = np.random.poisson(0.5 if projects > 1 else 0.2)
        if integrations > 0:
            prob += integrations * 0.04  # Each integration adds 4%
#-------------------------------------------------------------------------------------------------
        # Simulate a user’s average daily activity score (scale: 5–97)
        activity = np.random.randint(5, 98)
        if activity > 70:
            prob += 0.20  # Highly active
        elif activity > 45:
            prob += 0.10  # Moderately active
        elif activity > 20:
            prob += 0.03  # Low activity
#--------------------------------------------------------------------------------------------------
        # Add random noise to simulate real-world unpredictability and clip
        final_prob = np.clip(prob + np.random.normal(0, 0.08), 0.02, 0.98)

        # Simulate final conversion (1 = paid, 0 = not paid)
        converted = int(np.random.rand() < final_prob)

        # Append the user record to the data list
        data.append([
            user_id, projects, tasks, invites, integrations,
            activity, onboarding, channel, converted
        ])

    # Return the final data as a Pandas DataFrame
    return pd.DataFrame(data, columns=[
        'user_id', 'projects_created', 'tasks_assigned_to_others',
        'team_members_invited', 'integrations_used',
        'avg_daily_activity_score', 'completed_onboarding_tutorial',
        'marketing_channel', 'converted_to_paid'
    ])

df = generate_data()
print("Sample of Generated Data:")
print(df.head())
print(f"\nShape of generated data: {df.shape}")
print(f"Conversion Rate: {df['converted_to_paid'].mean():.2%}\n")
print("-" * 80 + "\n")
```

```python

# --- STEP 2.5: DATA CLEANING & VALIDATION ---
print("STEP 2.5: DATA CLEANING & VALIDATION")
print("Performing basic checks for data quality. In a real-world scenario, this step would be more involved.")

#---------------------------------------------------------------------------------------------------------------------------------------
# 1. Check for Missing Values
print("\n1. Checking for Missing Values:")
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0]) # Print only columns with missing values
if missing_values.sum() == 0:
    print("  No missing values found in the dataset. (As expected for simulated data)")
else:
    print("  Missing values found! Handling strategies would be needed (e.g., imputation, removal).")
    # Example: df.fillna(df.mean(), inplace=True) for numerical, or df.dropna(inplace=True)
#---------------------------------------------------------------------------------------------------------------------------------------
# 2. Check Data Types
print("\n2. Checking Data Types:")
print(df.dtypes)
# In a real scenario, you'd verify if these match expectations (e.g., numbers are numeric, categories are objects/categories).
# Example: df['projects_created'] = pd.to_numeric(df['projects_created'], errors='coerce')
#---------------------------------------------------------------------------------------------------------------------------------------

# 3. Check for Duplicates
print("\n3. Checking for Duplicate Rows (excluding user_id for perfect duplicates):")
# If user_id is unique, duplicates in other columns for different users are not an issue.
# We'd check for complete duplicate rows:
duplicate_rows = df.duplicated().sum()
if duplicate_rows > 0:
    print(f"  Found {duplicate_rows} duplicate rows. These might need to be removed.")
    # df.drop_duplicates(inplace=True)
else:
    print("  No complete duplicate rows found.")
#---------------------------------------------------------------------------------------------------------------------------------------
# 4. Basic Value Range Checks (Example for 'avg_daily_activity_score')
print("\n4. Basic Value Range Checks (Example for 'avg_daily_activity_score'):")
score_min = df['avg_daily_activity_score'].min()
score_max = df['avg_daily_activity_score'].max()
print(f"  'avg_daily_activity_score' ranges from {score_min} to {score_max}.")
if score_min < 0 or score_max > 100:
    print("  Warning: 'avg_daily_activity_score' is outside the expected 0-100 range. Investigation needed.")
# For categorical data like 'marketing_channel', check unique values:
print("\n  Unique values in 'marketing_channel':")
print(f"  {df['marketing_channel'].unique()}")
# Here you'd check for typos or inconsistent capitalization if it were real data.

#---------------------------------------------------------------------------------------------------------------------------------------
# 5. Outlier Detection and Handling (using IQR method)
print("\n5. Outlier Detection (using IQR method):")

# Define numerical features for outlier check (excluding binary 0/1 features like 'completed_onboarding_tutorial'
# as IQR might not be as meaningful for them, though it can still be applied)
numerical_cols_for_outliers = ['projects_created', 'tasks_assigned_to_others',
                               'team_members_invited', 'integrations_used',
                               'avg_daily_activity_score']

# Store original descriptive stats for comparison if we cap outliers
original_descriptives = df[numerical_cols_for_outliers].describe()

outliers_found_summary = {}

for col in numerical_cols_for_outliers:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    num_outliers = len(outliers)
    outliers_found_summary[col] = num_outliers

    if num_outliers > 0:
        print(f"  Found {num_outliers} outliers in '{col}' (values < {lower_bound:.2f} or > {upper_bound:.2f}).")

        # --- Optional: Outlier Handling (Capping/Winsorization) ---
        # Uncomment the following lines to apply capping
        print(f"    Capping outliers in '{col}' to range [{lower_bound:.2f}, {upper_bound:.2f}]...")
        df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
        df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])
        print(f"    '{col}' after capping: min={df[col].min():.2f}, max={df[col].max():.2f}")
        # --- End of Optional Capping ---
    else:
        print(f"  No outliers detected in '{col}' using the 1.5*IQR rule.")

print("\nOutlier Detection Summary (Number of outliers by 1.5*IQR rule):")
for col, count in outliers_found_summary.items():
    print(f"  - {col}: {count} outliers")

# If capping was applied, you might want to show new descriptive stats:
if any(val > 0 for val in outliers_found_summary.values()): # Check if any capping happened
  print("\nDescriptive statistics AFTER potential outlier capping:")
  print(df[numerical_cols_for_outliers].describe())
else:
  print("\nNo capping applied, descriptive statistics remain the same as original.")


print("\nData Cleaning & Validation Summary:")
if sum(outliers_found_summary.values()) > 0:
    print("  Basic data checks performed. Outliers were detected (and optionally handled by capping).")
else:
    print("  Basic data checks performed. No significant outliers detected by the IQR method.")



print("  For this simulated dataset, other checks (missing values, types, duplicates) pass as expected.")
print("  In real-world projects, this step is critical and often iterative with EDA.")
print("-" * 80 + "\n")


```

```python
# --- STEP 4: DATA PREPROCESSING ---
print("STEP 4: DATA PREPROCESSING")
X = df.drop(['user_id', 'converted_to_paid'], axis=1) # X → All the features (what the model uses to make a prediction).
y = df['converted_to_paid'] # y → The target variable (whether the user converted to paid or not).

categorical_features = ['marketing_channel'] # Categorical = Text/category column (like 'marketing_channel': Organic, Paid_Ad, etc.)
numerical_features = ['projects_created', 'tasks_assigned_to_others', 'team_members_invited',
                      'integrations_used', 'avg_daily_activity_score', 'completed_onboarding_tutorial'] # onboarding is 0/1, Numerical = All the number-based columns (project counts, score, etc.)

# Create preprocessing pipelines
numerical_pipeline = StandardScaler() # Simpler: directly use the scaler: For numerical features: We're scaling them so all numbers lie on the same scale (like bringing them to the same height before comparison).
categorical_pipeline = OneHotEncoder(handle_unknown='ignore', sparse_output=False) # sparse_output=False for easier handling: For categorical features: We convert them into numbers using One-Hot Encoding (turning categories into columns with 0s and 1s).

# Create a column transformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_pipeline, numerical_features),
        ('cat', categorical_pipeline, categorical_features)
    ],
    remainder='drop' # Drop any columns not specified
) # This tool automatically applies the right transformations (scaling or encoding) to the right columns.

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=SEED, stratify=y)

print(f"Training data shape: X_train: {X_train.shape}, y_train: {y_train.shape}")
print(f"Testing data shape: X_test: {X_test.shape}, y_test: {y_test.shape}")
print("-" * 80 + "\n")
```

## **STEP 5: MODEL TRAINING, EVALUATION & COMPARISON**

```python
# --- STEP 5: MODEL TRAINING, EVALUATION & COMPARISON
print("STEP 5: MODEL TRAINING, EVALUATION & COMPARISON (Using Default Hyperparameters)")

models = {
    "Logistic Regression": LogisticRegression(solver='liblinear', random_state=SEED),
    "K-Nearest Neighbors": KNeighborsClassifier(), # Default n_neighbors=5
    "Decision Tree": DecisionTreeClassifier(random_state=SEED), # Default parameters
    "Random Forest": RandomForestClassifier(random_state=SEED), # Default n_estimators=100
    "Gradient Boosting": GradientBoostingClassifier(random_state=SEED), # Default parameters
    "Support Vector Machine": SVC(probability=True, random_state=SEED), # Default C=1.0, kernel='rbf'
    "Gaussian Naive Bayes": GaussianNB()
}

results = {}
trained_pipelines = {} # Store trained model pipelines

for model_name, model_instance in models.items():
    print(f"Training {model_name}...")

    # Create a full pipeline with preprocessor and the model instance
    pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                               ('classifier', model_instance)])

    # Train the model
    pipeline.fit(X_train, y_train)
    trained_pipelines[model_name] = pipeline # Store for later use

    # Make predictions
    y_pred_test = pipeline.predict(X_test)
    y_proba_test = np.nan # Default if predict_proba is not available
    try:
        y_proba_test = pipeline.predict_proba(X_test)[:, 1]
        roc_auc = roc_auc_score(y_test, y_proba_test)
    except AttributeError: # Some models might not have predict_proba by default or if not applicable
        roc_auc = "N/A"
        print(f"  Note: ROC AUC not available for {model_name} with current setup.")


    # Evaluate
    results[model_name] = {
        "Test Accuracy": accuracy_score(y_test, y_pred_test),
        "Precision": precision_score(y_test, y_pred_test, zero_division=0),
        "Recall": recall_score(y_test, y_pred_test, zero_division=0),
        "F1-score": f1_score(y_test, y_pred_test, zero_division=0),
        "ROC AUC": roc_auc
    }
    print(f"  {model_name} - Test F1-score: {results[model_name]['F1-score']:.4f}, Test ROC AUC: {str(results[model_name]['ROC AUC'])}")


results_df = pd.DataFrame(results).T

# Sort by F1-score, handling 'N/A' in ROC AUC if any model didn't provide it
# Convert ROC AUC to numeric, coercing errors to NaN for sorting
results_df['ROC AUC_numeric'] = pd.to_numeric(results_df['ROC AUC'], errors='coerce')
results_df = results_df.sort_values(by="F1-score", ascending=False)


print("\nModel Performance Comparison (Test Set - Default Hyperparameters):")
# Displaying without the temporary numeric ROC AUC column
print(results_df[['Test Accuracy', 'Precision', 'Recall', 'F1-score', 'ROC AUC']])
print("-" * 80 + "\n")
```

# **STEP 6: SELECTION OF BEST MODEL**

```python
# --- STEP 6: SELECTION OF BEST MODEL (Simplified) ---
print("STEP 6: SELECTION OF BEST MODEL")
# Select based on F1-score as it's good for imbalanced classes
best_model_name = results_df['F1-score'].idxmax()
best_pipeline = trained_pipelines[best_model_name]
best_model_metrics = results_df.loc[best_model_name]

print(f"The best performing model (default params) based on F1-score is: **{best_model_name}**")
print("Metrics for the best model:")
print(best_model_metrics[['Test Accuracy', 'Precision', 'Recall', 'F1-score', 'ROC AUC']]) # Show selected metrics

# Simplified Feature Importance (if applicable)
feature_importance_df = None
if hasattr(best_pipeline.named_steps['classifier'], 'feature_importances_'): # Tree-based models
    try:
        transformer = best_pipeline.named_steps['preprocessor']
        transformed_feature_names = []
        # Numerical features names are direct
        transformed_feature_names.extend(numerical_features)
        # Categorical features names from OneHotEncoder
        ohe_categories = transformer.transformers_[1][1].categories_ # Get categories from OHE
        for i, cat_feature in enumerate(categorical_features):
            for category in ohe_categories[i]:
                transformed_feature_names.append(f"{cat_feature}_{category}")

        importances = best_pipeline.named_steps['classifier'].feature_importances_
        if len(transformed_feature_names) == len(importances):
            feature_importance_df = pd.DataFrame({'feature': transformed_feature_names, 'importance': importances})
            feature_importance_df = feature_importance_df.sort_values(by='importance', ascending=False)
            print("\nTop 5 Feature Importances for the best model:")
            print(feature_importance_df.head())
        else:
            print("\nMismatch in feature names and importances length. Skipping feature importance display.")

    except Exception as e:
        print(f"\nCould not retrieve feature importances for {best_model_name}: {e}")

elif best_model_name == "Logistic Regression":
    try:
        transformer = best_pipeline.named_steps['preprocessor']
        transformed_feature_names = []
        transformed_feature_names.extend(numerical_features)
        ohe_categories = transformer.transformers_[1][1].categories_
        for i, cat_feature in enumerate(categorical_features):
            for category in ohe_categories[i]:
                transformed_feature_names.append(f"{cat_feature}_{category}")

        coefficients = best_pipeline.named_steps['classifier'].coef_[0]
        if len(transformed_feature_names) == len(coefficients):
            feature_importance_df = pd.DataFrame({'feature': transformed_feature_names, 'importance_coeff': np.abs(coefficients)})
            feature_importance_df = feature_importance_df.sort_values(by='importance_coeff', ascending=False)
            print("\nTop 5 Feature Coefficients (Magnitude) for Logistic Regression:")
            print(feature_importance_df.head())
        else:
            print("\nMismatch in feature names and coefficients length for LR. Skipping display.")
    except Exception as e:
        print(f"\nCould not retrieve coefficients for Logistic Regression: {e}")

print("-" * 80 + "\n")
```
